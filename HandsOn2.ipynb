## Implenmentation for each algorithm

```python
# Insertion Sort
def insertionSort(a):
    for i in range(1, len(a)):
        # The current element
        key=a[i]
        j=i
        # Move elements before a[i], compare them with current element
        while j>0 and a[j-1]>key:
            a[j]=a[j-1] # Shift the element to the right
            j-=1
        a[j]=key # Insert the element at the correct position
    return a
```

```python
# Selection Sort
def selectionSort(a):
    for i in range(len(a)-1):
        # Assume the current element is the smallest
        minIndex=i
        # Find the element after a[i] and smaller than a[i]
        for j in range(i+1, len(a)):
            if a[j]<a[minIndex]:
                # If an element is smaller than minimum, swap
                minIndex=j
                a[i], a[minIndex]=a[minIndex], a[i]

    return a
```

```python
# Bubble Sort
def bubbleSort(a):
    for i in range (len(a)-1):
        for j in range(len(a)-1-i):
            # Compare the current element with the next
            if a[j]>a[j+1]:
                # Swap the elements if the current is larger
                a[j], a[j+1]=a[j+1], a[j]

    return a
```

```python
# test
arr=[1,3,54,6,2,8,34,9,4]
print(insertionSort(arr))
print(bubbleSort(arr))
print(selectionSort(arr))
```

## Selection Sort Correctness
1. Initialization: Before the first iteration, the subarray of sorted elements is empty, which trivially satisfies the invariant.
2. Maintenance: Assume that the invariant holds before the $i$-th iteration. During the $i$-th iteration, the algorithm selects the smallest element from the unsorted subarray and swaps it with the first element of the unsorted subarray. This ensures that the first (i+1) elements are sorted and the invariant holds.
3. Termination: As the loop terminates, the entire array is sorted, and the invariant guarantees that the array is correctly sorted.

## Benchmark the runtime of each algorithm
**Computer Information**
CPU: Apple M1 Pro, RAM: 16 GB

```python
from random import randint
from time import time
import matplotlib.pyplot as plt
sizes=[5, 10, 20, 40, 80, 160, 320, 640, 1280, 2560, 5120, 10240]
insertionTimes=[]
selectionTimes=[]
bubbleTimes=[]
for size in sizes:
    # Generate an array
    arr=[]
    for i in range(size):
        arr.append(randint(1, 100))

    # Sort the array, calculate the time
    timeStart=time()
    insertionSort(arr)
    timeEnd=time()
    # print(insertionTime)
    insertionTimes.append(timeEnd-timeStart)
    timeStart=time()
    selectionSort(arr)
    timeEnd=time()
    selectionTimes.append(timeEnd-timeStart)
    timeStart=time()
    bubbleSort(arr)
    timeEnd=time()
    bubbleTimes.append(timeEnd-timeStart)

plt.plot(sizes, insertionTimes, label='insertion')
plt.plot(sizes, selectionTimes, label='selection')
plt.plot(sizes, bubbleTimes, label='bubble')
# Set x-axis to logarithmic scale
plt.xscale('log')
plt.xticks(sizes, sizes)  # Ensure ticks are placed at the correct positions
plt.legend()
plt.title('Benchmark the runtime of each algorithm')
plt.xlabel('Array Size')
plt.ylabel('Time')
```
