## 1 The runtime of each algorithm

$$T(n)=1+\sum_{i=1}^{n}\sum_{j=1}^{n}1=n^2+1$$

## 2 Plot Curve

```python
import matplotlib.pyplot as plt
n=[i for i in range(1, 200)]
Tn=[i**2+1 for i in n]
# print(Tn)
plt.plot(n, Tn, label='time')
plt.legend()
plt.title('Runtime of the Algorithm')
plt.xlabel('n')
plt.ylabel('Tn')
```

## 3 & 4 Find polynomials & $n_0$

Upper Bounds: $2n^2$, Lower Bounds: $n^2$

For any $n\geq 1$, i.e. $n_0=1$, $n^2\leq T(n)\leq 2n^2$, therefore,

$T(n)$ is $O(n^2)$, $\Theta(n^2)$, and $\Omega(n^2)$.

A plot is shown below.

```python
n=[i for i in range(1, 10)] # To show clearly, a smaller n is used
Tn=[i**2+1 for i in n]
lower=[i**2 for i in n]
upper=[2*(i**2) for i in n]
# print(Tn)
plt.plot(n, Tn, label='time')
plt.plot(n, lower, label='lower')
plt.plot(n, upper, label='upper')
plt.legend()
plt.title('Upper and Lower Bound of the Runtime')
plt.xlabel('n')
plt.ylabel('Tn')
```

## 5 Modified Algorithm

The runtime of the modified algorithm is 

$$T(n)=2+\sum_{i=1}^{n}\sum_{j=1}^{n}2=2(1+n^2)$$

Therefore the runtime increases, and it affect the results in #1.

```python
n=[i for i in range(1, 200)] # To show clearly, a smaller n is used
Tn=[i**2+1 for i in n]
newTn=[2*(1+i**2) for i in n]
plt.plot(n, Tn, label='Time')
plt.plot(n, newTn, label='Updated Time')
plt.legend()
plt.title('Runtime of the updated Algorithm')
plt.xlabel('n')
plt.ylabel('Tn')
```

## 7 Implement Merge Sort

```python
def merge_sort(arr):
    if len(arr)>1:
        mid=len(arr)//2  # Finding the mid of the array
        L=arr[:mid]  # Dividing the elements into 2 halves
        R=arr[mid:]

        print(f"Dividing: {arr} into {L} and {R}")

        merge_sort(L)  # Sorting the first half
        merge_sort(R)  # Sorting the second half

        i=j=k=0

        # Copy data to temp arrays L[] and R[]
        while i<len(L) and j<len(R):
            if L[i]<R[j]:
                arr[k]=L[i]
                i+=1
            else:
                arr[k]=R[j]
                j+=1
            k+=1

        # Check if any element was left
        while i<len(L):
            arr[k]=L[i]
            i+=1
            k+=1

        while j<len(R):
            arr[k]=R[j]
            j+=1
            k+=1

        print(f"Merging: {arr}")

merge_sort([5,2,4,7,1,3,2,6])
```
